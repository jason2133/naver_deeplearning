# Image Handling 이미지 만지기

```python
# 패키지 불러오기
# Numpy, OS, Scipy, Matplotlib

import numpy as np
# numpy

import os
# os

from scipy.misc import imread, imresize
# scipy
# imread : 이미지 읽어요, imresize : 이미지 사이즈 재조정
# imageread, imageresize

import matplotlib.pyplot as plt
%matplotlib inline
# matplotlib
```

## Print Current Folder

```python
# getcwd : 현재 디렉토리 위치를 알려준다

cwd = os.getcwd()
print("Current Folder is %s" % (cwd))
```

## Print Function

```python
# Type과 Shape 출력!
# type(), img.shape,
# type() 데이터 타입
# img.shape, 이미지 모양

def print_typeshape(img):
    
    print("Type is %s" % (type(img)))
    # 데이터 타입을 알려줘요!
    
    print("Shape is %s" % (img.shape,))
    # 이미지의 모양을 알려줘요!
    # 뒤에 , 있는거 조심!
    # img.shape,
```

## Load an Image

```python
# 고양이

cat = imread("./data/cat.jpg")
print_typeshape(cat)
```

## Plot Loaded Image

```python
plt.figure(figsize = (10, 8)) # 그림 사이즈
plt.imshow(cat) # 어떤 그림을 보여주지?
plt.title("Original CAT") # 그림 위에 제목
plt.draw() # 그려줘요
```

```python
cat
# 정수로 이루어져 있음
# 0 ~ 255 사이
```

## Load + Cast to Float? 소수로 하기?

```python
# Load
cat2 = imread("./data/cat.jpg").astype(np.float) # np.float : 소수로 처리한다!
print_typeshape(cat2)

# Plot
plt.figure(figsize = (10, 8))
plt.imshow(cat2)
plt.title("Original Image with imread.astype(np.float)")
plt.draw()
```

0에서 1 사이의 값을 가지는 소수 처리가 되어서 그림이 이상하게 보인다

## Load + Cast to Float + Normalize to 0~1 정상화 시키기

```python
# Load
cat3 = imread("./data/cat.jpg").astype(np.float)
print_typeshape(cat3)

# Plot
plt.figure(figsize = (10, 8))
plt.imshow(cat3/255.)
# 그림 정상화시키려면 여기다가 /255. 를 붙여아 함!

plt.title("Original Image with imread.astype(np.float)/255.")
plt.draw()
```

## Resize

```python
# Resize
catsmall = imresize(cat, [100, 100])
# 100, 100 사이즈로 사진을 재조정한다

print_typeshape(catsmall)

# plot
plt.figure(1)
plt.imshow(catsmall)
plt.title("Resized Image")
plt.draw()
```

```python
# Resize
catsmall2 = imresize(cat, [200, 100])
# 200, 100으로 사이즈 재조정

print_typeshape(catsmall2)

# plot
plt.figure(1)
plt.imshow(catsmall2)
plt.title("Resized Image 2")
plt.draw()
```

## GrayScale

```python
def rgb2gray(rgb):
    if len(rgb.shape) is 3 : # channel이 3
        return np.dot(rgb[..., :3], [0.299, 0.587, 0.114])
    else:
        print("Current Image if GRAY!")
        return rgb
```

```python
catsmallgray = rgb2gray(catsmall)
print("size of catsmallgray is %s" % (catsmallgray.shape,)) # shape,
print("type of catsmallgray is %s" % (type(catsmallgray))) # type()
```

```python
# 흑백
# cmap = plt.get_cmap 얘가 있어야 함!
# 행렬을 다루기 때문에 column이라는 값을 통해서 각각의 값을 다른 색깔로 표현해주어야 함

plt.imshow(catsmallgray, cmap = plt.get_cmap("gray"))
plt.title("[imshow] Gray Image")
plt.colorbar()
plt.draw()
```

```python
# 제트
# 적외선

plt.imshow(catsmallgray, cmap = plt.get_cmap("jet"))
plt.title("[imshow] Jet Image")
plt.colorbar()
plt.draw()
```

## Reshape 모양을 바꾼다

### Matrix to Vector

```python
# Matrix to Vector

catrowvec = np.reshape(catsmallgray, (1, -1))
# 이거 꿀팁
# 1, -1 하면 -1이 10000이 됨
# 애초에 -1을 붙이면 두개를 곱한 값이 10000으로 해줌
# 여기서는 1, 10000이 된다는 소리겠지

print("size of catsmallgray is %s" % (catrowvec.shape,))
print("type of catsmallgray is %s" % (type(catrowvec)))
```

```python
plt.imshow(catrowvec, cmap=plt.get_cmap("gray"))
plt.title("Reshaped Matrix - Catrowvec")
plt.colorbar()
plt.draw()
```

### Vector to Matrix

```python
# Vector to Matrix

catmatrix = np.reshape(catrowvec, (100, -1))
# 여기도 마찬가지로 -1이 있으면 두개를 곱하면 10000
# 따라서 여기는 100, 100이 됨!

print("size of catmatrix is %s" % (catmatrix.shape,))
print("type of catmatrix is %s" % (type(catmatrix)))
```

```python
plt.imshow(catmatrix, cmap=plt.get_cmap("gray"))
plt.title("Reshaped Matrix - CatMatrix")
plt.colorbar()
plt.draw()
```

# Load From Folder

## First, See what's inside the Folder

```python
cwd = os.getcwd()
path = cwd + '/data/'
flist = os.listdir(path)
print("[%d] files are in [%s]" % (len(flist), path))
```

## List Up Files inside the Folder

```python
# enumerate 활용

for i, f in enumerate(flist):
    print("[%d]th file is [%s]" % (i, f))
```

## Load Images with Approximate Extensions

```python
valid_exts = [".jpg", ".gif", ".png", ".tga", ".jpeg"] # 이런 류의 파일들만 허용된다
imgs = []
names = []

for f in flist:
    ext = os.path.splitext(f)[1]
    if ext.lower() not in valid_exts: # 목록에 없으면 소문자 처리해라
        continue
    # Load and Append to IMGS List
    fullpath = os.path.join(path, f)
    # join 붙인다
    imgs.append(imread(fullpath))
    # append 추가시킨다
    
    # Append Names
    names.append(os.path.splitext(f)[0])
```

## Plot Loaded Images

```python
for img, name in zip(imgs, names):
    # 번호가 아니라 그 첫번째 부분 두번째 부분 이렇게 나눌 때 Zip를 사용한다!
    plt.imshow(img)
    plt.title(name)
    plt.show()
```
